# **Funções**

```python
# Escopo de Variáveis

# Variável Local
def funcao_local():
    variavel_local = 10
    print(variavel_local)

funcao_local()
# print(variavel_local) gera erro
```

```python
# Escopo de Variáveis

# Variável Global
variavel_global = 10

def funcao_global():
    global variavel_global
    variavel_global += 1
    print(variavel_global)

print(variavel_global)
funcao_global()
print(variavel_global)
```
